## 2: Plot 12 most dangerous cities

```python
city_freq = kill["city"].value_counts().head(12)
city_freq
```

```python
x=kill["city"].value_counts().head(12).index
y=kill["city"].value_counts().head(12).values
fig, ax =plt.subplots(figsize=(8,5))
fig.suptitle('Most dangerous cities',y=0.93,fontsize=15,color="blue");
ax = sns.barplot(x=x,y=y,data=kill)
plt.xticks(rotation=45);

plt.show()
```

```python
fig, ax =plt.subplots(figsize=(8,5))
sns.countplot(x="city",data=kill,order=kill['city'].value_counts().head(12).index)
fig.suptitle('Most dangerous cities',y=0.93,fontsize=15,color="blue");
plt.xticks(rotation=45);
```

## 3: Plot 20 most dangerous states

```python
state_freq = kill.state.value_counts().head(20)
state_freq
```

```python
x=kill["state"].value_counts().head(20).index
y=kill["state"].value_counts().head(20).values
fig, ax =plt.subplots(figsize=(8,5))
fig.suptitle('Most dangerous states',y=0.93,fontsize=15,color="blue");
ax = sns.barplot(x=x,y=y,data=kill)
plt.xticks(rotation=45);

plt.show()
```

```python
fig, ax =plt.subplots(figsize=(8,5))
sns.countplot(x="state",data=kill,order=kill['state'].value_counts().head(20).index)
fig.suptitle('Most dangerous states',y=0.93,fontsize=15,color="blue");
plt.xticks(rotation=45);
```
